```python
# Data Cleaning


df.drop_duplicates(inplace=True) #Drop for duplicate values
print(df.duplicated().sum())
df.fillna("NoValue")


display(df)
```

```python
#1.Which country has the most number of players (score :1)

#Data Manipulation

#Counting the country

country_counts = df['Nationality'].value_counts()

print(country_counts)

# Getting the country with the maximum count

most_players_country = country_counts.idxmax()

print("\n\nCountry with the most players:", most_players_country,"\nCount:",country_counts.max())
```

Insight

```python
#2.Plot a bar chart of 5 top countries with the most number of players. (score :1)

from matplotlib import pyplot as plt
import numpy as np


country_counts = df['Nationality'].value_counts()

# Select the top 5 countries
top_countries = country_counts.head(5)

# Plotting the bar chart
#plt.figure(figsize=(10, 6))
top_countries.plot(kind='bar', color='skyblue')
plt.title('Top 5 Countries with Most Players')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--')
plt.tight_layout() # To Assure elements and values are never overlaped 
plt.show()
```

Insight

```python
#3.Which player has the highest salary? (score :1)

#highest_salary_player = df.loc[df['Release Clause'].idxmax()]['Name']

df=pd.read_csv("C:\\Users\\rajan\\fifa_data.csv")

df = df[df['Wage'] != '€0']

df['Wage'] = df['Wage'].str.replace('[\€]', '', regex=True)
df['Wage'] = df['Wage'].str[:-1].astype(float)
highest_pay = df['Wage'].max()


index_max = df['Wage'].idxmax()

# Get the name of the player with the maximum wage
player_name= df.loc[index_max, 'Name']

print("Highest Salary\nPlayer: ", player_name,'\nSalary: €',highest_pay,"K")

#sorted_fifa_data = df.sort_values(by='Wage',ascending=False)

#print(sorted_fifa_data[["Name","Wage"]].head(20))





```

Insight
3.Highest Salary is for L. Messi and Salary is €565.0K

```python
#4.Plot a histogram to get the salary range of the players. (score :1)


from matplotlib import pyplot as plt
import numpy as np

#df=pd.read_csv("C:\\Users\\rajan\\fifa_data.csv")

plt.hist(df['Wage'])
plt.title('Salary Distribution of FIFA Players')
plt.xlabel('Salary(K)')
plt.ylabel('No Of Players')
plt.grid(True)
plt.grid(axis='y', linestyle='--')
plt.tight_layout() # To Assure elements and values are never overlaped 

plt.show()

print("Most the players salary is in between 0K-50K")


```

```python
count = df['Club'].value_counts().idxmax()
player_count = df['Club'].value_counts().max()
print("Club with the most number of players:", count)
print("Number of players:", player_count)
```

```python
# Calculate the player counts for each club
club_player_counts = df['Club'].value_counts()

# Find the maximum count
max_count = club_player_counts.max()
# Filter clubs with the maximum count
clubs_with_max_count = club_player_counts[club_player_counts == max_count]

print("Clubs with the most number of players ({} players each):".format(max_count))
print(clubs_with_max_count)
print("Total Number of Clubs with most number of players (33 players each) ",clubs_with_max_count.count())
```

```python
#7.Which foot is most preferred by the players?Draw a bar chart for preferred foot (score :1)

import matplotlib.pyplot as plt

# Count the occurrences of each foot preference
foot_preference_counts = df['Preferred Foot'].value_counts()
print(foot_preference_counts)
# Plot a bar chart
foot_preference_counts.plot(kind='bar', color='pink')

# Add title and labels
plt.title('Preferred Foot of Players')
plt.xlabel('Foot Preference')
plt.ylabel('Number of Players')
plt.legend()
plt.grid(axis='y', linestyle='--')
plt.tight_layout() # To Assure elements and values are never overlaped 

plt.show()
```

Insight
7.Most of the players prefer their right foot . 13948 players prefer right foot and 4211
 Players prefer the left foot.
From the bar chart we will understand that around 14000 prefers right foot and above 4000 and less than 4500 prefer left foot
